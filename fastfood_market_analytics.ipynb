```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Carregar o arquivo top_50_fast_food_US.csv
file_path = '/content/top_50_fast_food_US-1.csv'
df = pd.read_csv(file_path)

print("## Primeiras 5 linhas do DataFrame:\n")
# 2. Mostrar as 5 primeiras linhas
print(df.head())
print("\n" + "-" * 50 + "\n")

print("## Verificação de valores ausentes:\n")
# 3. Verificar valores ausentes
print(df.isnull().sum())
print("\n" + "-" * 50 + "\n")

# 4. Criar um gráfico de barras horizontal com as 10 maiores empresas em vendas totais (sales_in_millions_2019)
# Converter 'sales_in_millions_2019' para tipo numérico, lidando com possíveis erros
df['sales_in_millions_2019'] = pd.to_numeric(df['sales_in_millions_2019'], errors='coerce')

# Remover linhas onde 'sales_in_millions_2019' se tornou NaN após a conversão
df_cleaned = df.dropna(subset=['sales_in_millions_2019'])

top_10_companies = df_cleaned.nlargest(10, 'sales_in_millions_2019')

plt.figure(figsize=(12, 8))
sns.barplot(x='sales_in_millions_2019', y='company', data=top_10_companies, palette='viridis')

# 5. Adicionar título e rótulos
plt.title('Top 10 Empresas de Fast Food nos EUA por Vendas (2019)', fontsize=16)
plt.xlabel('Vendas (em milhões de dólares, 2019)', fontsize=12)
plt.ylabel('Empresa', fontsize=12)
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

print("\n" + "-" * 50 + "\n")
print("## Análise do Gráfico:\n")
# 6. Explicar o que o gráfico revela sobre o mercado de fast food nos EUA em 2019.
print("O gráfico de barras horizontal acima mostra as 10 maiores empresas de fast food nos EUA com base em suas vendas em milhões de dólares em 2019.\n")
print("**O que o gráfico revela:**\n")
print("1.  **Dominância de Poucas Grandes Redes:** Fica evidente que algumas poucas redes de fast food dominam significativamente o mercado em termos de vendas. A diferença entre a primeira colocada e as subsequentes pode ser bastante expressiva, indicando uma alta concentração de mercado.\n")
print("2.  **Líderes de Mercado Claros:** Empresas como McDonald's, Starbucks e Chick-fil-A (se aparecerem entre os primeiros) consistentemente demonstram vendas muito superiores a outras, solidificando suas posições como líderes incontestáveis.\n")
print("3.  **Diversidade de Ofertas:** Embora sejam todas 'fast food', as empresas na lista frequentemente representam diferentes segmentos (hambúrgueres, café, frango, etc.), mostrando que o 'fast food' abrange uma vasta gama de categorias de alimentos e bebidas que atraem diferentes bases de consumidores.\n")
print("4.  **Mercado Maduro e Competitivo:** A existência de um 'top 10' com vendas na casa dos bilhões de dólares sugere um mercado de fast food extremamente maduro e competitivo, onde as marcas estabelecidas mantêm uma forte presença e fidelidade do cliente.\n")
print("Em resumo, o gráfico destaca a estrutura oligopolista do mercado de fast food nos EUA em 2019, com poucas mega-corporações controlando a maior parte das vendas, mesmo em um setor tão diversificado.")
```

### Análise de Contexto Detalhada

```python
# 1. Quantas redes estão no dataset?
num_redes = len(df_cleaned)
print(f"Número total de redes no dataset: {num_redes}\n")

# 2. Quais categorias estão presentes?
categorias_presentes = df_cleaned['category'].unique()
print(f"Categorias presentes no dataset: {', '.join(categorias_presentes)}\n")

# 3. Qual o período dos dados? Qual a fonte?
print("Período dos dados: 2019 (inferido pelas colunas de vendas como 'sales_in_millions_2019')\n")
print("Fonte dos dados: top_50_fast_food_US-1.csv\n")

# 4. Quais são as 5 maiores redes em vendas totais?
top_5_vendas = df_cleaned.nlargest(5, 'sales_in_millions_2019')
print("As 5 maiores redes em vendas totais (2019):\n")
for index, row in top_5_vendas.iterrows():
    print(f"- {row['company']}: ${row['sales_in_millions_2019']:,} milhões")
print("\n")

# 5. Quais são as 5 redes com maior número total de unidades?
top_5_unidades = df_cleaned.nlargest(5, 'total_units_2019')
print("As 5 redes com maior número total de unidades (2019):\n")
for index, row in top_5_unidades.iterrows():
    print(f"- {row['company']}: {row['total_units_2019']:,} unidades")
```

### Análise de Eficiência

```python
# 1. Identificar as 5 redes mais eficientes (maior venda por unidade)
top_5_eficientes = df_cleaned.nlargest(5, 'sales_per_unit_thousands_2019')
print("As 5 redes mais eficientes por vendas por unidade (2019):\n")
for index, row in top_5_eficientes.iterrows():
    print(f"- {row['company']}: ${row['sales_per_unit_thousands_2019']:,} mil por unidade")
print("\n")

# 2. Identificar as 5 redes menos eficientes (menor venda por unidade)
top_5_ineficientes = df_cleaned.nsmallest(5, 'sales_per_unit_thousands_2019')
print("As 5 redes menos eficientes por vendas por unidade (2019):\n")
for index, row in top_5_ineficientes.iterrows():
    print(f"- {row['company']}: ${row['sales_per_unit_thousands_2019']:,} mil por unidade")
print("\n")

# 3. Pergunta de negócio: Redes com muitas unidades são sempre mais eficientes por loja? Justifique.
print("**Redes com muitas unidades são sempre mais eficientes por loja? Justifique.**\n")
print("Não necessariamente. Embora redes maiores possam se beneficiar de economias de escala em termos de poder de compra e marketing, um grande número de unidades não garante automaticamente uma maior eficiência por loja.\n")
print("**Justificativas:**\n")
print("    *   **Saturação de Mercado:** Em alguns casos, um número excessivo de unidades em uma área pode levar à canibalização das vendas, diminuindo a receita média por loja.\n")
print("    *   **Gestão de Qualidade e Operacional:** Manter a consistência na qualidade do serviço e do produto em muitas unidades pode ser um desafio, e falhas nesse aspecto podem impactar negativamente a eficiência individual das lojas.\n")
print("    *   **Localização:** A localização estratégica de poucas unidades pode gerar maior volume de vendas por loja do que um grande número de unidades mal localizadas.\n")
print("    *   **Modelo de Negócio:** Algumas redes podem focar em um modelo de alto volume e baixo custo (e.g., fast food tradicional), enquanto outras podem priorizar um valor de ticket médio mais alto com menos transações (e.g., cafeterias premium ou restaurantes de 'fast casual'), o que se reflete nas vendas por unidade.\n")
print("O gráfico de dispersão a seguir nos ajudará a visualizar essa relação, mostrando se há uma correlação clara entre o número de unidades e a eficiência de vendas por unidade.")

# 4. Gráfico de dispersão (scatter plot) com total_units_2019 no eixo X e sales_per_unit_thousands_2019 no eixo Y.
plt.figure(figsize=(14, 9))
sns.scatterplot(x='total_units_2019', y='sales_per_unit_thousands_2019', data=df_cleaned, hue='company', size='sales_in_millions_2019', sizes=(50, 1000), legend='brief')

# Destaque a rede mais eficiente e a maior rede em unidades.
most_efficient = df_cleaned.loc[df_cleaned['sales_per_unit_thousands_2019'].idxmax()]
largest_units = df_cleaned.loc[df_cleaned['total_units_2019'].idxmax()]

# Adicionar anotações para os destaques
plt.annotate(most_efficient['company'],
             (most_efficient['total_units_2019'], most_efficient['sales_per_unit_thousands_2019']),
             textcoords="offset points", xytext=(5,5), ha='left', color='red', weight='bold')
plt.annotate(largest_units['company'],
             (largest_units['total_units_2019'], largest_units['sales_per_unit_thousands_2019']),
             textcoords="offset points", xytext=(5,-10), ha='left', color='blue', weight='bold')

plt.title('Total de Unidades vs. Vendas por Unidade (2019)', fontsize=16)
plt.xlabel('Total de Unidades (2019)', fontsize=12)
plt.ylabel('Vendas por Unidade (milhares de dólares, 2019)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.xscale('log') # Usar escala logarítmica no eixo X para melhor visualização de grandes variações
plt.yscale('log') # Usar escala logarítmica no eixo Y para melhor visualização
plt.tight_layout()
plt.show()
```

### Comparação entre Modelo Franqueado vs. Unidades Próprias

```python
# Calcular percentual de unidades franqueadas e próprias
df_cleaned['percent_franchised'] = (df_cleaned['franchised_units_2019'] / df_cleaned['total_units_2019']) * 100
df_cleaned['percent_company_owned'] = (df_cleaned['company_owned_units_2019'] / df_cleaned['total_units_2019']) * 100

# Tratar casos onde total_units_2019 é 0 para evitar NaNs ou inf
df_cleaned.loc[df_cleaned['total_units_2019'] == 0, 'percent_franchised'] = 0
df_cleaned.loc[df_cleaned['total_units_2019'] == 0, 'percent_company_owned'] = 0

print("## Percentual de unidades franqueadas e próprias por rede (primeiras 5 linhas):\n")
print(df_cleaned[['company', 'percent_franchised', 'percent_company_owned']].head())
print("\n" + "-" * 50 + "\n")

# Existe relação entre alto percentual de franquias e crescimento de unidades (unit_change_from_2018)?
print("**Existe relação entre alto percentual de franquias e crescimento de unidades (unit_change_from_2018)?**\n")
correlation = df_cleaned['percent_franchised'].corr(df_cleaned['unit_change_from_2018'])
print(f"A correlação entre o percentual de unidades franqueadas e a mudança de unidades é: {correlation:.2f}\n")
print("Uma correlação positiva (próxima de 1) indicaria que redes com mais franquias tendem a crescer mais em número de unidades, enquanto uma correlação negativa (próxima de -1) indicaria o oposto. Uma correlação próxima de zero sugere pouca ou nenhuma relação linear direta.\n")
print("\n" + "-" * 50 + "\n")

# Quais redes são 100% franqueadas? E 100% próprias?
print("**Redes 100% Franqueadas:**\n")
franchised_100 = df_cleaned[df_cleaned['percent_franchised'] == 100]
if not franchised_100.empty:
    for company in franchised_100['company']:
        print(f"- {company}")
else:
    print("Nenhuma rede encontrada com 100% de unidades franqueadas.")
print("\n")

print("**Redes 100% Próprias:**\n")
company_owned_100 = df_cleaned[df_cleaned['percent_company_owned'] == 100]
if not company_owned_100.empty:
    for company in company_owned_100['company']:
        print(f"- {company}")
else:
    print("Nenhuma rede encontrada com 100% de unidades próprias.")
print("\n" + "-" * 50 + "\n")

# O modelo de negócios influencia a eficiência (vendas por unidade)?
# Compare a média de vendas por unidade entre redes majoritariamente franqueadas (>80% franquias) e majoritariamente próprias (>80% próprias).

# Criar grupos
majority_franchised = df_cleaned[df_cleaned['percent_franchised'] > 80]
majority_company_owned = df_cleaned[df_cleaned['percent_company_owned'] > 80]

mean_sales_franchised = majority_franchised['sales_per_unit_thousands_2019'].mean()
mean_sales_company_owned = majority_company_owned['sales_per_unit_thousands_2019'].mean()

print("**O modelo de negócios influencia a eficiência (vendas por unidade)?**\n")
print(f"Média de vendas por unidade para redes majoritariamente franqueadas (>80%): ${mean_sales_franchised:,.2f} mil\n")
print(f"Média de vendas por unidade para redes majoritariamente próprias (>80%): ${mean_sales_company_owned:,.2f} mil\n")

if mean_sales_franchised > mean_sales_company_owned:
    print("Redes majoritariamente franqueadas parecem ter, em média, maior eficiência de vendas por unidade neste dataset.")
elif mean_sales_company_owned > mean_sales_franchised:
    print("Redes majoritariamente próprias parecem ter, em média, maior eficiência de vendas por unidade neste dataset.")
else:
    print("A média de vendas por unidade é semelhante entre redes majoritariamente franqueadas e próprias neste dataset.")
print("\n" + "-" * 50 + "\n")
```

### Gráficos para Comparação entre Modelo Franqueado vs. Unidades Próprias

```python
# Gráfico de barras empilhadas mostrando a composição (franquia vs própria) para as 10 maiores redes em vendas.

top_10_sales = df_cleaned.nlargest(10, 'sales_in_millions_2019')

# Preparar dados para o gráfico empilhado
plot_data = top_10_sales[['company', 'percent_franchised', 'percent_company_owned']].set_index('company')

plt.figure(figsize=(14, 8))
plot_data.plot(kind='barh', stacked=True, color=['#1f77b4', '#ff7f0e'])

plt.title('Composição de Unidades (Franqueadas vs. Próprias) para as Top 10 Redes em Vendas (2019)', fontsize=16)
plt.xlabel('Percentual de Unidades (%)', fontsize=12)
plt.ylabel('Empresa', fontsize=12)
plt.xticks(rotation=0)
plt.legend(['Franqueadas', 'Próprias'], title='Tipo de Unidade')
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

print("\n" + "-" * 50 + "\n")

# Boxplot comparando sales_per_unit_thousands_2019 entre os dois grupos
# Criar uma coluna 'ownership_type' para categorizar as redes
def get_ownership_type(row):
    if row['percent_franchised'] > 80:
        return 'Majoritariamente Franqueada'
    elif row['percent_company_owned'] > 80:
        return 'Majoritariamente Própria'
    else:
        return 'Misto'

df_cleaned['ownership_type'] = df_cleaned.apply(get_ownership_type, axis=1)

plt.figure(figsize=(10, 7))
sns.boxplot(x='ownership_type', y='sales_per_unit_thousands_2019', data=df_cleaned[df_cleaned['ownership_type'] != 'Misto'], palette='coolwarm')

plt.title('Vendas por Unidade por Tipo de Modelo de Negócio (2019)', fontsize=16)
plt.xlabel('Tipo de Modelo de Negócio', fontsize=12)
plt.ylabel('Vendas por Unidade (milhares de dólares)', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

### Segmentação por Categoria de Alimento

```python
# Agrupar os dados por 'category' e calcular as métricas
category_analysis = df_cleaned.groupby('category').agg(
    total_sales=('sales_in_millions_2019', 'sum'),
    avg_sales_per_unit=('sales_per_unit_thousands_2019', 'mean'),
    total_units=('total_units_2019', 'sum'),
    avg_unit_change=('unit_change_from_2018', 'mean')
).reset_index()

print("## Análise por Categoria de Alimento:\n")
print(category_analysis)
print("\n" + "-" * 50 + "\n")

# Responder às perguntas:

# 1. Qual categoria tem o maior faturamento total?
max_sales_category = category_analysis.loc[category_analysis['total_sales'].idxmax()]
print(f"**Categoria com o maior faturamento total:** {max_sales_category['category']} (${max_sales_category['total_sales']:,} milhões)\n")

# 2. Qual categoria tem a maior eficiência média (vendas por unidade)?
max_efficiency_category = category_analysis.loc[category_analysis['avg_sales_per_unit'].idxmax()]
print(f"**Categoria com a maior eficiência média (vendas por unidade):** {max_efficiency_category['category']} (${max_efficiency_category['avg_sales_per_unit']:.2f} mil por unidade)\n")

# 3. Em qual categoria as redes mais cresceram em número de unidades (2018–2019)?
max_growth_category = category_analysis.loc[category_analysis['avg_unit_change'].idxmax()]
print(f"**Categoria com maior crescimento médio de unidades:** {max_growth_category['category']} ({max_growth_category['avg_unit_change']:.2f} unidades)\n")
print("\n" + "-" * 50 + "\n")
```

### Gráficos por Categoria de Alimento

```python
# Gráfico de barras para vendas totais por categoria.
plt.figure(figsize=(12, 7))
sns.barplot(x='total_sales', y='category', data=category_analysis.sort_values('total_sales', ascending=False), palette='magma')
plt.title('Vendas Totais por Categoria de Alimento (2019)', fontsize=16)
plt.xlabel('Vendas Totais (milhões de dólares)', fontsize=12)
plt.ylabel('Categoria', fontsize=12)
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

# Gráfico de barras para média de vendas por unidade por categoria.
plt.figure(figsize=(12, 7))
sns.barplot(x='avg_sales_per_unit', y='category', data=category_analysis.sort_values('avg_sales_per_unit', ascending=False), palette='viridis')
plt.title('Média de Vendas por Unidade por Categoria de Alimento (2019)', fontsize=16)
plt.xlabel('Média de Vendas por Unidade (milhares de dólares)', fontsize=12)
plt.ylabel('Categoria', fontsize=12)
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

# Gráfico de barras para a variação média de unidades por categoria.
plt.figure(figsize=(12, 7))
sns.barplot(x='avg_unit_change', y='category', data=category_analysis.sort_values('avg_unit_change', ascending=False), palette='plasma')
plt.title('Variação Média de Unidades por Categoria (2018-2019)', fontsize=16)
plt.xlabel('Variação Média de Unidades', fontsize=12)
plt.ylabel('Categoria', fontsize=12)
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

# 📊 Conclusões Finais

---

## Principais Insights

### 1️⃣ Liderança de Mercado
As maiores redes de fast food concentram uma parcela significativa do faturamento do setor, demonstrando forte presença de marca e elevado alcance nacional.

### 2️⃣ Eficiência Operacional
Foi observado que possuir mais unidades não garante necessariamente maior eficiência em vendas por loja. Algumas redes com menor número de estabelecimentos apresentaram desempenho superior por unidade.

### 3️⃣ Modelo de Expansão
O sistema de franquias mostrou-se predominante entre as maiores empresas do mercado, funcionando como importante estratégia de crescimento e escalabilidade.

### 4️⃣ Diferenças Entre Categorias
As categorias analisadas apresentaram comportamentos distintos em faturamento, crescimento e eficiência operacional, evidenciando características específicas de cada segmento.

---

## Conclusão Geral

A análise demonstra que o sucesso de uma rede de fast food não depende apenas do número de unidades ou do faturamento total. Fatores como eficiência operacional, estratégia de expansão e posicionamento competitivo exercem papel fundamental no desempenho das empresas. Dessa forma, decisões gerenciais devem considerar múltiplos indicadores para avaliar o verdadeiro potencial de crescimento e rentabilidade de uma rede.
